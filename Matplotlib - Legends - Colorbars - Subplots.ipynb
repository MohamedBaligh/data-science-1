```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

cities = pd.read_csv('california_cities.csv')
```

```python
cities.head()
```

```python
lat,lon = cities['latd'], cities['longd']
```

```python
population, area = cities['population_total'],cities['area_total_km2']
```

```python
plt.scatter(lon, lat, c = np.log10(population), cmap='viridis', s=area, alpha=0.5)
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar(label='log10(population)')
```

```python
# load images of the digits 0 through 5 and visualize several of them
from sklearn.datasets import load_digits
digits = load_digits(n_class=6)
```

```python
for i,j in enumerate([11,22,33,44]):
    print(i,j)
```

```python
fig, ax = plt.subplots(4,4, figsize=(6,6))
# ax is a 4 X 4 storage, ax.flat maps it into 16 size of 1 d array
for i,axi in enumerate(ax.flat):
    axi.imshow(digits.images[i], cmap='inferno')
    axi.set(xticks=[], yticks=[]) # disable axis marking
    
```
