```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
```

```python
ax = plt.axes()
```

```python
fig = plt.figure() # create a fig object
ax = plt.axes() # create axes
x = np.linspace(0,10,1000) # generate points
ax.plot(x, np.sin(x)) # plot sin curve
ax.plot(x, np.cos(x),) # plot cos curve
```

### Line Colors & Styles

```python
plt.plot(x, np.sin(x), color='red')
```

```python
plt.plot(x, np.sin(x), color='#91487B')
```

```python
plt.plot(x, np.sin(x), color='#91487B', linestyle='-.')
```

```python
plt.plot(x, np.sin(x),'-.r')
```

### Axes Limits

```python
plt.plot(x, np.sin(x))
plt.xlim(-5,10)
plt.ylim(0,20)
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
plt.title('My matplot graph')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot( x, np.sin(x), label='sin(x)')
plt.plot( x, np.cos(x), label='cos(x)')
plt.axis('equal')
# plotting the legend
plt.legend()
```
